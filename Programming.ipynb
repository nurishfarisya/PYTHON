```python
# Full Program
# Function to add a product
def add_product():
    # Determine the next available Product ID
    with open("products.txt", "r") as file:
        lines = file.readlines()
        if lines:
            last_line = lines[-1].strip()
            last_id = int(last_line.split('|')[0].strip()[1:])
            next_id = f"P{str(last_id + 1).zfill(3)}"
        else:
            next_id = "P001"
    print(f"Next available Product ID: {next_id}")

    # Get a valid Product ID from the user
    while True:
        product_id = input(f"Enter Product ID (or press Enter to use {next_id}): ").strip()
        if not product_id:  # Use suggested ID if input is empty
            product_id = next_id
        if len(product_id) == 4 and product_id[0] == 'P' and product_id[1:].isdigit():
            if any(product_id == line.split('|')[0].strip() for line in lines):
                print(f"Error: Product ID {product_id} already exists. Please try again.")
            else:
                break
        else:
            print(f"Error: Product ID must be in the format PXXX, where X is a digit (e.g., {next_id}). Please try again.")

    # Get a valid Product Name
    while True:
        name = input("Enter Product Name: ").strip()
        if not name:
            print("Error: Product Name cannot be empty. Please try again.")
        else:
            break

    # Get a valid Product Description
    while True:
        description = input("Enter Product Description: ").strip()
        if not description:
            print("Error: Product Description cannot be empty. Please try again.")
        else:
            break

    # Get a valid Product Price
    while True:
        try:
            price = float(input("Enter Product Price: ").strip())
            if price <= 0:
                print("Error: Price must be greater than 0. Please try again.")
            else:
                break
        except ValueError:
            print("Error: Please enter a valid number for price.")

    # Append the new product to the file
    with open("products.txt", "a") as file:
        file.write(f"\n{product_id} | {name} | {description} | {price:.2f}")
    print(f"\nProduct {product_id} added successfully!")

def update_product():
    
    # Read products from the file
    with open("products.txt", "r") as file:
        lines = file.readlines()

    # Display all products
    print("\nCurrent Products:")
    for line in lines:
        print(line.strip())

    # Ask for the Product ID to update
    while True:
        product_id = input("\nEnter the Product ID to update: ").strip()
        if product_id:  # Ensure input is not empty
            break
        print("Product ID cannot be empty. Please try again.")

    product_found = False

    for i, line in enumerate(lines):
        # Check if the product ID matches
        if line.split('|')[0].strip() == product_id:
            product_found = True

            # Display current product details
            print(f"\nCurrent details for Product {product_id}:")
            print(line.strip())

            # Get new details for the product
            new_name = input("Enter new Product Name (or press Enter to keep unchanged): ").strip()
            new_description = input("Enter new Product Description (or press Enter to keep unchanged): ").strip()

            # Get a valid new price
            while True:
                new_price = input("Enter new Product Price (or press Enter to keep unchanged): ").strip()
                if not new_price:  # Keep current price if input is empty
                    new_price = line.split('|')[3].strip()
                    break
                try:
                    new_price = float(new_price)
                    if new_price <= 0:
                        print("Price must be greater than 0. Please try again.")
                    else:
                        new_price = f"{new_price:.2f}"  # Format price to two decimal places
                        break
                except ValueError:
                    print("Please enter a valid number for price.")

            # Update the product details
            updated_line = f"{product_id} | {new_name or line.split('|')[1].strip()} | {new_description or line.split('|')[2].strip()} | {new_price}\n"
            lines[i] = updated_line

            # Write the updated details back to the file
            with open("products.txt", "w") as file:
                file.writelines(lines)

            print(f"\nProduct {product_id} updated successfully!")
            return

    if not product_found:
        print(f"\nError: Product ID {product_id} not found. Please try again.")
        update_product()  # Retry without restarting

# Function to add a supplier
def add_supplier():
    # Determine the next available Supplier ID
    with open("suppliers.txt", "r") as file:
        lines = file.readlines()
        if lines:
            last_line = lines[-1].strip()
            last_id = int(last_line.split('|')[0].strip()[1:])
            next_id = f"S{str(last_id + 1).zfill(3)}"
        else:
            next_id = "S001"
    print(f"Next available Supplier ID: {next_id}")

    # Get a valid Supplier ID from the user
    while True:
        supplier_id = input(f"Enter Supplier ID (or press Enter to use {next_id}): ").strip()
        if not supplier_id:  # Use suggested ID if input is empty
            supplier_id = next_id
        if len(supplier_id) == 4 and supplier_id[0] == 'S' and supplier_id[1:].isdigit():
            if any(supplier_id == line.split('|')[0].strip() for line in lines):
                print(f"Error: Supplier ID {supplier_id} already exists. Please try again.")
            else:
                break
        else:
            print(f"Error: Supplier ID must be in the format SXXX, where X is a digit (e.g., {next_id}). Please try again.")

    # Get a valid Supplier Name
    while True:
        name = input("Enter Supplier Name: ").strip()
        if not name:
            print("Error: Supplier Name cannot be empty. Please try again.")
        else:
            break

    # Get a valid Email Address
    while True:
        email = input("Enter Supplier Email: ").strip()
        if '@' in email and '.' in email:
            break
        else:
            print("Error: Invalid email format. Please enter a valid email (e.g., example@domain.com)")

    # Get a valid Phone Number
    print("\nPhone number format example: 012-123-1234")
    while True:
        phone = input("Enter Phone Number (XXX-XXX-XXXX): ").strip()
        if (
            len(phone) == 12 and
            phone[:3].isdigit() and
            phone[3] == '-' and
            phone[4:7].isdigit() and
            phone[7] == '-' and
            phone[8:].isdigit()
        ):
            break
        else:
            print("Error: Phone must be in format XXX-XXX-XXXX (e.g., 012-123-1234). Please try again.")

    # Append the new supplier to the file
    with open("suppliers.txt", "a") as file:
          file.write(f"\n{supplier_id} | {name} | {email} | {phone}")
    print(f"\nSupplier {supplier_id} added successfully!")

# Facilitate the process of placing an order for a specific product
def place_order():

    # Determine the next available Order ID based on existing data
    with open("orders.txt", "r") as file:
        lines = file.readlines()
        if lines:
            last_line = lines[-1]
            last_id = int(last_line.split('|')[0].strip()[1:])
            next_id = f"O{str(last_id + 1).zfill(3)}"
            print(f"Next available Order ID: {next_id}")
        else:
            next_id = "O001"
            print("No existing orders. Starting with ID: O001")

    # Get order ID from user
    while True:
        order_id = input(f"Enter Order ID (or press Enter to use {next_id}): ").strip()
        if not order_id:  # Use suggested ID if input is empty
            order_id = next_id
        if len(order_id) == 4 and order_id[0] == 'O' and order_id[1:].isdigit():
            with open("orders.txt", "r") as file:
                if any(order_id == line.split('|')[0].strip() for line in file):
                    print(f"Error: Order ID {order_id} already exists. Please use a different ID.")
                    continue
            break
        else:
            print(f"Error: Order ID must be in the format OXXX, where X is a digit (e.g., {next_id}). Please try again.")

    # Get product ID from user
    while True:
        product_id = input("Enter Product ID (e.g., P001): ").strip()
        if len(product_id) == 4 and product_id[0] == 'P' and product_id[1:].isdigit():
            break
        else:
            print("Error: Product ID must be in the format PXXX, where X is a digit (e.g., P001). Please try again.")

    # Get quantity from user
    while True:
        try:
            quantity = int(input("Enter Quantity: ").strip())
            if quantity > 0:
                break
            else:
                print("Error: Quantity must be a positive number. Please try again.")
        except ValueError:
            print("Error: Please enter a valid number.")

    # Get current date
    from datetime import datetime
    current_date = datetime.now().strftime("%d.%m.%Y")

    # Append the order to the file
    with open("orders.txt", "a") as file:
        file.write(f"\n{order_id} | {product_id} | {quantity} | {current_date}")
        print(f"\nOrder {order_id} placed successfully!")

# Display the current inventory levels for all products
def view_inventory():
   # Display header showing column titles for product information
   print("\nProduct ID | Product Name | Description | Price(RM)")
   # Create visual separator line using dashes for better readability
   print('-' * 60)

   try:
       # Open products.txt file in read mode and process its contents
       with open("products.txt", "r") as file:
           # Read all lines from file at once and store in list
           file_contents = file.readlines()
           
           # Check if file contains any product entries
           if file_contents:
               # Loop through each line and display product information
               for line in file_contents:
                   # Remove extra whitespace/newlines before printing
                   print(line.strip())
           else:
               # Display message if no products are found in file
               print("No products found in inventory.")
               
   # Handle any input/output errors that occur when reading file
   except IOError as e:
       # Print the specific error message
       print(e)
       # Display user-friendly error message
       print("Cannot read from products file.")

# Function to generate a report
# Low Stock Report
def low_stock_report():
        print("\n=== Low Stock Items Report === ")
        try:
            # Read all products
            with open("products.txt", "r") as file:
                product_lines = file.readlines()

            # Read all orders to calculate current stock
            product_orders = {}
            with open("orders.txt", "r") as file:
                order_lines = file.readlines()
                for order in order_lines:
                    # Order format: OrderID | ProductID | Quantity | Date
                    data = order.strip().split('|')
                    product_id = data[1].strip()
                    quantity = int(data[2].strip())
                    product_orders[product_id] = product_orders.get(product_id, 0) + quantity

            if product_lines:
                print("Product ID | Name | Current Stock | Status")
                print("-" * 60)

                # Process each product
                for line in product_lines:
                    data = line.strip().split('|')
                    product_id = data[0].strip()
                    name = data[1].strip()

                    # Calculate current stock (initial stock - orders)
                    initial_stock = 20  # Set initial stock level
                    orders = product_orders.get(product_id, 0)
                    current_stock = initial_stock - orders

                    # Determine status
                    if current_stock < 15:
                        status = "LOW STOCK"
                    else:
                        status = "OK"

                    print(f"{product_id} | {name} | {current_stock} | {status}")
            else:
                print("No products found in inventory.")

        except Exception as e:
            print("Error! There was an issue processing the data in required file.")
            
# Products Sales Report
def product_sales_report():
        print("\n=== Product Sales Report === ")
        try:
            # Open and read two files
            with open("products.txt", "r") as file, open("orders.txt", "r") as orders_file:
                lines = file.readlines()
                orders = orders_file.readlines()

                # Create a dictionary for product data
                product_data = {}
                for line in lines:
                    data = line.strip().split('|')
                    product_id = data[0].strip()
                    name = data[1].strip()
                    price = float(data[3].strip())
                    product_data[product_id] = {"name": name, "price": price}

                # Create a dictionary for sales data
                product_sales = {}
                for order in orders:
                    data = order.strip().split('|')
                    product_id = data[1].strip()
                    quantity = int(data[2].strip())
                    product_sales[product_id] = product_sales.get(product_id, 0) + quantity

                # Generate the sales report
                if product_sales:
                    print("Product ID | Name | Total Sold | Total Sales (RM) ")
                    print("-" * 60)
                    # Sort by Product ID before printing
                    for product_id in sorted(product_sales.keys()):
                        product = product_data.get(product_id, {})
                        name = product.get("name", "Unknown")
                        price = product.get("price", 0)
                        total_sold = product_sales[product_id]
                        total = total_sold * price
                        print(f"{product_id} | {name} | {total_sold} | {total:.2f}")
                else:
                    print("No sales data found. ")

        except IOError:
            print("Error! There was an issue processing the data in required file. ")

# Supplier Orders Report
def supplier_orders_report():
        print("\n=== Supplier Orders Report === ")
        try:
            # First get supplier information
            supplier_names = {}
            with open("suppliers.txt", "r") as file:
                for line in file:
                    data = line.strip().split('|')
                    supplier_id = data[0].strip()  # e.g., S001
                    name = data[1].strip()         # e.g., Fresh Farms
                    supplier_names[supplier_id] = name

            # Now get products and their suppliers
            product_suppliers = {}
            with open("products.txt", "r") as file:
                for line in file:
                    data = line.strip().split('|')
                    product_id = data[0].strip()   # e.g., P001
                    # Map each product to its supplier
                    # Example: P001 maps to S001, P002 to S001, etc.
                    supplier_id = f"S{product_id[1:]}"  # Convert P001 to S001
                    product_suppliers[product_id] = supplier_id

            # Calculate orders per supplier
            supplier_orders = {}
            with open("orders.txt", "r") as file:
                for line in file:
                    data = line.strip().split('|')
                    product_id = data[1].strip()   # Get product ID
                    quantity = int(data[2].strip()) # Get quantity

                    # Get supplier for this product
                    supplier_id = product_suppliers.get(product_id)
                    if supplier_id:
                        supplier_orders[supplier_id] = supplier_orders.get(supplier_id, 0) + quantity

            # Generate report
            if supplier_orders:
                print("Supplier ID | Name | Total Orders")
                print("-" * 60)
                # Sort by supplier ID
                for supplier_id in sorted(supplier_orders.keys()):
                    name = supplier_names.get(supplier_id, "Unknown")
                    total = supplier_orders[supplier_id]
                    print(f"{supplier_id} | {name} | {total}")
            else:
                print("No supplier orders found.")

        except Exception as e:
            print(f"Error! There was an issue processing the data in required file. {str(e)}")

def generate_reports():
    while True:
        # Show the selection for report
        print("\nReport Generation Menu: ")
        print("[1] Low Stock Items Report")
        print("[2] Product Sales Report")
        print("[3] Supplier Orders Report")
        print("[4] Return to Main Menu")

        # Get the choice from user
        try:
            choice = int(input("Enter your choice (1-4): ").strip())
            match choice:
                case 1:
                    low_stock_report()
                case 2:
                    product_sales_report()
                case 3:
                    supplier_orders_report()
                case 4:
                    print("Returning to Main Menu...")
                    break
                case _:
                    # Handle invalid option
                    print("Invalid option! Please select between 1-4")

        except ValueError:
            # Handle non-integer input
            print('\nERROR: Please enter numbers only!')

# The main program that combines all functions
def main_program():
    running = True    # Boolean flag for program control
    
    while running:
        # Display menu
        print("\nInventory Management System")
        print("[1] Add a new product")
        print("[2] Update product details")
        print("[3] Add a new supplier")
        print("[4] Place an order")
        print("[5] View inventory")
        print("[6] Generate reports")
        print("[7] Exit program")
        
        # Handle input validation
        try:
            choice = int(input("Please enter your choice: "))
            
            # Process user selection using match-case for better readability
            match choice:
                case 1:
                    add_product()  # Call function to add a new product
                case 2:
                    update_product()  # Call function to update product details
                case 3:
                    add_supplier()  # Call function to add a new supplier
                case 4:
                    place_order()  # Call function to place an order
                case 5:
                    view_inventory()  # Call function to view current inventory
                case 6:
                    generate_reports()  # Call function to generate reports
                case 7:
                    break   # Just break out of the loop directly
                case _:
                    print("\nInvalid option! Please select between 1-7")  # Handle invalid option
                    
        except ValueError:
            print("\nERROR: Please enter numbers only!")  # Handle non-integer input
        
        # Only ask to continue if not exiting
        if running:
            while True:
                continue_input = input("\nDo you wish to continue using the system? [y/n]: ").lower()
                if continue_input in ['y', 'n']:
                    break  # Exit the loop if valid input is received
                print("Invalid input! Please enter y or n only")  # Handle invalid input
            
            if continue_input == 'n':
                running = False  # Set running to False if user chooses not to continue
                
    print("\nThank you for using the Inventory Management System! Goodbye!")  # Exit message

# Start the program
main_program()
```
